# What's the Cheapest Alcohol in BC?

Being a student is expensive so I need liquor that's cheap. Luckily, the Government of BC has kindly open sourced the all sorts of data about BC Liquor. This data was last updated in June 2016.

The data can be found here: http://bit.ly/2cccpx2

```python
import pandas as pd
```

```python
liquor_df = pd.read_csv(r'bcliquorstoreproductpricelistjune2016.csv')
liquor_df.drop(['PRODUCT_SKU_NO', 'PRODUCT_BASE_UPC_NO', 'SWEETNESS_CODE'], axis=1, inplace=True)
```

```python
liquor_df.head()
```

To find the cheapest drink, divide the price by the product of the Alcohol percent, container amount and number of containers.

```python
liquor_df['COST_RATIO'] = liquor_df['PRODUCT_PRICE']/((liquor_df['PRODUCT_ALCOHOL_PERCENT']/100)*liquor_df['PRODUCT_LITRES_PER_CONTAINER']*liquor_df['PRD_CONTAINER_PER_SELL_UNIT'])
```

```python
liquor_df.sort_values('COST_RATIO', ascending=True, inplace=True)
liquor_df.head()
```

There you have it, the cheapest alcohol is Andre's Medium Dry Fortified Wine. The second cheapest alcohol is...Andre's Medium Dry Fortified Wine. If you look closely, you can see that item 1 and 2 are different sizes. Every thrifty shopper knows that it's better to buy in bulk. Let's remove all sizes of the product except the one with the lowest cost ratio.

```python
liquor_df.reset_index()
liquor_df.drop_duplicates(subset=['PRODUCT_LONG_NAME'], inplace=True)
liquor_df.head()
```
